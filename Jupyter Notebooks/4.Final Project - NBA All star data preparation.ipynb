# Final Project - NBA All star data preparation

## Import libraries

```python
import pandas as pd
import numpy as np
```

## reading both df, all star and college players

```python
# reading allstar data file from csv

all_star_df = pd.read_csv('all_star_rosters.csv', index_col = 0)
all_star_df.head()
```

```python
all_star_df.shape
```

```python
# reading college data clean file from csv

college_players = pd.read_csv('college_data.csv', index_col = 0)
college_players.head()
```

## get only drafted players from college

```python

drafted_df=college_players.loc[college_players['drafted'] == 'Yes']
```

```python
drafted_df.head()
```

## feature engineering

### create all star column in college players data drame based on names in allstar data frame

```python
drafted_df['Allstar'] = np.where(drafted_df.player_name.isin(all_star_df.Name), 'Yes', 'No')
```

```python
drafted_df.reset_index()
```

```python
drafted_df['Allstar'].unique()
```

```python
# verify number of allstar players match with allstar df
drafted_df['Allstar'].value_counts()
```

```python
drafted_df.info()
```

```python
drafted_df=drafted_df.reset_index(drop=True)
```

```python
drafted_df.info()
```

```python
# showing allstar players, non college, overseas players or players drafted before 2008 won't appear
drafted_df.loc[drafted_df['Allstar'] == 'Yes']
```
